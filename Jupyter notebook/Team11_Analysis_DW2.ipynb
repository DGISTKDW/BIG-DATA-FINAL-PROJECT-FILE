## Blood Press EDA with alcohol and age

```python
#2009
bp_2009 = GJ_2009[['alcohol','age','b_press_contract','b_press_relax']]
bp_2009_0=bp_2009[bp_2009.alcohol==0]
bp_2009_1=bp_2009[bp_2009.alcohol==1]
bp_2009_mean_0 = bp_2009_0.groupby('age').mean()
bp_2009_mean_1 = bp_2009_1.groupby('age').mean()
bp_2009_mean_0 = bp_2009_mean_0.rename(columns={'b_press_contract': 'No_alcohol_contract', 'b_press_relax': 'No_alcohol_relax'})
bp_2009_mean_1 = bp_2009_mean_1.rename(columns={'b_press_contract': 'Alcohol_contract', 'b_press_relax': 'Alcohol_relax'})
bp_2009_mean_0 = bp_2009_mean_0.drop('alcohol', axis = 1)
bp_2009_mean_1 = bp_2009_mean_1.drop('alcohol', axis = 1)
c_bp_mean_2009_smoke = pd.concat([bp_2009_mean_0, bp_2009_mean_1], axis=1, join='inner')
c_bp_mean_2009_smoke.plot(linestyle = '-', marker = 'o', figsize = (10, 5))
```
