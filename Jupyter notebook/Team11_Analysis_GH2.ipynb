## (A) Blood press

### (1) blood press by age

```python
#2005
bp_2005=GJ_2005[['sex','age','b_press_contract','b_press_relax']]
bp_2005_male=bp_2005[bp_2005.sex==1]
bp_2005_female=bp_2005[bp_2005.sex==2]
bp_mean_2005_male=bp_2005_male.groupby('age').mean()
bp_mean_2005_female=bp_2005_female.groupby('age').mean()
bp_mean_2005_male=bp_mean_2005_male.rename(columns={'b_press_contract':'male_contract'})
bp_mean_2005_male=bp_mean_2005_male.rename(columns={'b_press_relax':'male_relax'})
bp_mean_2005_female=bp_mean_2005_female.rename(columns={'b_press_contract':'female_contract'})
bp_mean_2005_female=bp_mean_2005_female.rename(columns={'b_press_relax':'female_relax'})
merge_data_2005=pd.merge(bp_mean_2005_male,bp_mean_2005_female, left_index=True, right_index=True)
merge_data_2005=merge_data_2005[['male_contract','female_contract','male_relax','female_relax']]
merge_data_2005=merge_data_2005.iloc[:12]
merge_data_2005
```

```python
merge_data_2005=merge_data_2005.reset_index()
merge_data_2005.plot(x='age',y=['male_contract','female_contract','male_relax','female_relax'],linestyle='-',marker='o',figsize=((10,5)))
```
