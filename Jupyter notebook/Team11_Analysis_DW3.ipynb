## EDA with two columns

```python
GJ_2009 = pd.read_csv('NHIS_OPEN_GJ_2009.CSV')
GJ_2009 = GJ_2009[GJ_2009['age'] < 13]
GJ_2009 = GJ_2009[GJ_2009['height'] <= 180]
GJ_2009 = GJ_2009[GJ_2009['height'] >= 145]
GJ_2009 = GJ_2009[GJ_2009['weight'] <= 90]
GJ_2009 = GJ_2009[GJ_2009['weight'] >= 45]
GJ_2009.head()
```

## Blood Press EDA with age and sex

```python
#2009
bp_2009 = GJ_2009[['sex','age','b_press_contract','b_press_relax']]
bp_2009_male=bp_2009[bp_2009.sex==1]
bp_2009_female=bp_2009[bp_2009.sex==2]
bp_mean_2009_male=bp_2009_male.groupby('age').mean()
bp_mean_2009_female=bp_2009_female.groupby('age').mean()
bp_mean_2009_male=bp_mean_2009_male.rename(columns={'b_press_contract':'male_contract','b_press_relax':'male_relax'})
bp_mean_2009_female=bp_mean_2009_female.rename(columns={'b_press_contract':'female_contract', 'b_press_relax':'female_relax'})
bp_mean_2009_male = bp_mean_2009_male.drop('sex', axis = 1)
bp_mean_2009_female = bp_mean_2009_female.drop('sex', axis = 1)
c_bp_mean_2009_sex = pd.concat([bp_mean_2009_male, bp_mean_2009_female], axis=1, join='inner')
c_bp_mean_2009_sex.plot(linestyle = '-', marker = 'o', figsize = (10, 5))
```
