```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style("whitegrid")
from pylab import figure, axes, pie, title, savefig
```

```python
JR_2013 = pd.read_csv('NHIS_OPEN_T20_2013.csv', encoding='CP949')
JR_2014 = pd.read_csv('NHIS_OPEN_T20_2014.csv', encoding='CP949')
JR_2015 = pd.read_csv('NHIS_OPEN_T20_2015.csv', encoding='CP949')
JR_2010_2015 = pd.concat([JR_2013, JR_2014, JR_2015])
JR_2010_2015.head()
```

```python
JR_2010_2015 = JR_2010_2015[['요양개시일자', '주상병코드','시도코드']]
JR_2010_2015.head()
```

```python
JR_2010_2015['요양개시일자'] = pd.to_datetime(JR_2010_2015['요양개시일자'], format = '%Y%m%d', errors = 'coerce')
JR_2010_2015 = JR_2010_2015[JR_2010_2015['시도코드'] == 41]
J0_JR_2010_2015_code1 = JR_2010_2015[JR_2010_2015['주상병코드'].str.match('J0')]
J2_JR_2010_2015_code1 = JR_2010_2015[JR_2010_2015['주상병코드'].str.match('J2')]
J3_JR_2010_2015_code1 = JR_2010_2015[JR_2010_2015['주상병코드'].str.match('J3')]
J0_JR_2010_2015_code1_ct = J0_JR_2010_2015_code1['요양개시일자'].groupby(J0_JR_2010_2015_code1.요양개시일자.dt.month).agg('count')
J2_JR_2010_2015_code1_ct = J2_JR_2010_2015_code1['요양개시일자'].groupby(J2_JR_2010_2015_code1.요양개시일자.dt.month).agg('count')
J3_JR_2010_2015_code1_ct = J3_JR_2010_2015_code1['요양개시일자'].groupby(J3_JR_2010_2015_code1.요양개시일자.dt.month).agg('count')
```

```python
J0_JR_2010_2015_code1_ct = pd.DataFrame(J0_JR_2010_2015_code1_ct)
J2_JR_2010_2015_code1_ct = pd.DataFrame(J2_JR_2010_2015_code1_ct)
J3_JR_2010_2015_code1_ct = pd.DataFrame(J3_JR_2010_2015_code1_ct)
J0_JR_2010_2015_code1_ct = J0_JR_2010_2015_code1_ct.rename(columns= {'요양개시일자': 'Total_patients'})
J0_JR_2010_2015_code1_ct = J0_JR_2010_2015_code1_ct.reset_index()
J0_JR_2010_2015_code1_ct = J0_JR_2010_2015_code1_ct.rename(columns = {'요양개시일자': 'Month'})
J2_JR_2010_2015_code1_ct = J2_JR_2010_2015_code1_ct.rename(columns= {'요양개시일자': 'Total_patients'})
J2_JR_2010_2015_code1_ct = J2_JR_2010_2015_code1_ct.reset_index()
J2_JR_2010_2015_code1_ct = J2_JR_2010_2015_code1_ct.rename(columns = {'요양개시일자': 'Month'})
J3_JR_2010_2015_code1_ct = J3_JR_2010_2015_code1_ct.rename(columns= {'요양개시일자': 'Total_patients'})
J3_JR_2010_2015_code1_ct = J3_JR_2010_2015_code1_ct.reset_index()
J3_JR_2010_2015_code1_ct = J3_JR_2010_2015_code1_ct.rename(columns = {'요양개시일자': 'Month'})
```

```python
J_JR_2010_2015_code1_ct = pd.concat([J0_JR_2010_2015_code1_ct, J2_JR_2010_2015_code1_ct['Total_patients'], J3_JR_2010_2015_code1_ct['Total_patients']], axis = 1)
J_JR_2010_2015_code1_ct['First_Total_patients']= J_JR_2010_2015_code1_ct.loc[0:11].sum(axis=1)
J_JR_2010_2015_code1_ct = J_JR_2010_2015_code1_ct[['Month', 'First_Total_patients']]
J_JR_2010_2015_code1_ct
```

```python
f, ax = plt.subplots(figsize = (10, 15))

# plot the total treat_kind 1
sns.barplot(x = 'Month', y = 'First_Total_patients', data = J_JR_2010_2015_code1_ct)
plt.savefig('감기환자 월별 이용 인원.png', dpi=1000)
```
